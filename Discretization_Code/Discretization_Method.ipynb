```python
# Import pandas library / Importa a biblioteca pandas
import pandas as pd

# Import pandas library / Importa a biblioteca pandas
import pandas as pd

# Load data from the Excel file into a DataFrame / Carrega os dados do arquivo Excel em um DataFrame
file_path = '/content/cancer.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame / Exibe as primeiras linhas do DataFrame
display(df.head())
```

```python
# Split the 'radius_mean' column into 10 bins 📊/ Dividir a coluna 'radius_mean'em 10 faixas

bins = pd.qcut(df['radius_mean'], 10)

bins
```

```python
# Check with the value_counts method – it shows how many elements are in each group 🔎 Verificar coi metodo 'value_counts'- mostra quantos lelementos tem or grupoo


bins.value_counts()
```

```python
# Coluna 'radius_mean'é agrupadsa pelo metodo groupby

grupos = df['radius_mean'].groupby(bins)
```

```python
# obten a media de cada faixa (grupo

medias = grupos.mean()
medias
```

```python
#🇺🇸 **Discretization** – The process of taking a set of continuous data and splitting it into ranges (intervals), assigning a label to each range.

#**Apply** – The `apply` method takes a function as a parameter, applies that function to all elements, and returns a `pandas.Series` containing the results.
#➡️ In this case, `apply` will return the mean of each interval.

# ---


### 🇧🇷 **Discretização / Discretization** – Processo de pegar um conjunto de dados contínuos e dividi-lo em faixas (intervalos), atribuindo um rótulo para cada faixa.

# **Apply** – O método `apply` recebe como parâmetro uma função, aplica essa função a todos os elementos e retorna um `pandas.Series` contendo os resultados.
# ➡️ Neste caso, `apply` retornará a média de cada intervalo.




Radiu_mean2 = bins.apply(lambda a : medias[a])
```

```python
# Depois, a coluna 'radius_mean' é atualizada

df.radius_mean =  Radiu_mean2
df
```

```python
# Load data from the Excel file into a DataFrame again to get the original numeric column / Carrega os dados do arquivo Excel em um DataFrame novamente para obter a coluna numérica original
file_path = '/content/cancer.csv'
df_original = pd.read_csv(file_path)

# Divide the original 'radius_mean' column into 10 quantile-based bins / Dividir a coluna 'radius_mean' original em 10 faixas baseadas em quantis
bins_median = pd.qcut(df_original['radius_mean'], 10)

# Display the resulting bins / Exibir as faixas resultantes
display(bins_median.head())
```

## Calculate medians for each bin

### Subtask:
Calculate the median of 'radius_mean' for each of the created bins. / Calcular a mediana de 'radius_mean' para cada uma das faixas criadas.


**Reasoning**:
Group the original 'radius_mean' column by the created bins and calculate the median for each group, then display the result.



```python
# Group the original 'radius_mean' column from df_original by the bins_median series 📊/ Agrupar a coluna 'radius_mean' original do df_original pela série bins_median.
grupos_median = df_original['radius_mean'].groupby(bins_median)

# Calculate the median for each group using the .median() method 🔎 Calcular a mediana para cada grupo usando o método .median().
medians_median = grupos_median.median()

# Store the calculated medians in a variable named medians_median and display them / Armazenar as medianas calculadas em uma variável chamada medians_median e exibi-las.
display(medians_median)
```

## Apply medians to the column

### Subtask:
Replace the original 'radius_mean' values with the calculated median of the bin they belong to. / Substituir os valores originais de 'radius_mean' pela mediana calculada da faixa a que pertencem.


```python
# Replace the original 'radius_mean' values with the calculated median of the bin they belong to. / Substituir os valores originais de 'radius_mean' pela mediana calculada da faixa a que pertencem.
df['radius_mean'] = bins_median.apply(lambda x: medians_median[x])

# Display the updated DataFrame with the new 'radius_mean' values. / Exibir o DataFrame atualizado com os novos valores de 'radius_mean'.
display(df.head())
```

## Display the updated dataframe

### Subtask:
Show the first few rows of the DataFrame with the updated 'radius_mean' column. / Mostrar as primeiras linhas do DataFrame com a coluna 'radius_mean' atualizada.


```python
# Display the first few rows of the DataFrame with the updated 'radius_mean' column. / Mostrar as primeiras linhas do DataFrame com a coluna 'radius_mean' atualizada.
display(df.head())
```
